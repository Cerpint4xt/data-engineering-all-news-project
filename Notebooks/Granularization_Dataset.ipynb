```python
# importing necessary library
import pandas as pd
```

```python
# Creating dtypes and parse dates for good reading on the dataset
news_dtypes = {
    'year': pd.Int64Dtype(),
    'month': pd.Int64Dtype(),
    'day': pd.Int64Dtype(),
    'author': str,
    'title': str,
    'article': str,
    'url': str,
    'section': str,
    'publication': str
}
parse_dates = ['date']
```

Notice that the folders where the data is read and written are manually created with the year name

```python
def extract_csv_news(year, month, dataset):
  ''''extracting form the dataset loaded from hugging face to a csv file by year and month'''
  df_news = pd.read_csv('/content/drive/MyDrive/news_dataset/news_dataset_{}.csv'.format(dataset), dtype=news_dtypes , parse_dates=parse_dates)
  df_news.drop(columns=['Unnamed: 0','idx', 'article_idx'], inplace=True)
  df_news_year = df_news[df_news.year==year]
  df_news_year_month = df_news_year[df_news_year.month==month]
  df_news_year_month.to_csv('/content/news_data_year/{}/news_data_{}_{}_{}.csv'.format(year, year, month, dataset), index=False)
  return df_news_year_month
```

```python
# giving paramenters to process
year_list =[2020]
```

```python
# Running function to create month aggregation for each year.
for k in year_list:
  for j in (range(1, 13)):
    for i in (range(1,28)):
      extract_csv_news(k, j, i)
```

```python
# Looping to save each csv file in the corresponding file
for k in year_list:
  for j in range(1,13):
    df_by_year_month = pd.DataFrame()
    for i in range(1, 28):
      df_aux = pd.read_csv('/content/news_data_year/{}/news_data_{}_{}_{}.csv'.format(k, k, j, i))
      df_by_year_month = pd.concat([df_by_year_month, df_aux])
    #df_by_year_month.to_csv('/content/news_data/{}/news_data_{}_{:02d}.csv.gz'.format(k, k, j), index=False, compression='gzip')
    df_by_year_month.to_csv('/content/drive/MyDrive/news_dataset_compress/{}/news_data_{}_{:02d}.csv.gz'.format(k, k, j), index=False, compression='gzip')
```
